#Week 2, Exploratiry Data Analysis
In this notebook, we use Pandas and Matplotlib to load the Titanic dateset, explore basic statistics and visualize the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading the dataset by URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# We display the first 5 lines to get a visual overview of the data structure.
df.head()
```

```python
# General statistics for all numeric columns (count, average, min/max)
print("General statistics:")
display(df.describe())

# We calculate the mean and median for the 'Age' column.
mean_age = df['Age'].mean()
median_age = df["Age"].median()

print(f"\nAverage age of passengers: {mean_age:.2f} age")
print(f"\nMedian age of passengers: {median_age:.2f} age")
```

```python
# Set the chart size
plt.figure(figsize=(8,5))

# We're plotting a histogram. The .dropna() method is needed to ignore empty age values.
plt.hist(df["Age"].dropna(), bins = 20, color = 'skyblue', edgecolor = 'black')

# Adding captions to make the graph clearer
plt.title('Age distribution of Titanic passengers')
plt.xlabel('Age')
plt.ylabel('Number of passengers')

# Drowing
plt.show()
```
