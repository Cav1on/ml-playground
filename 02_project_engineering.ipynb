#Week 3: Code and project structure engineering

In this module, we'll move beyond simply writing code in cells to creating the structure of a real ML project.
We'll create folders and our first automated data processing script.

```python
import os

# Create a professional folder structure
folders = ['data','notebooks','src','models']

for folder in folders:
  if not os.path.exists(folder):
    os.makedirs(folder)
    print(f"Folder created: {folder}")
```

```python
%%writefile src/preprocess.py
import pandas as pd

def clean_titanic_data(url):
  """Function for basic data cleaning"""
  df = pd.read_csv(url)
  # We remove the unnecessary, leaving only what is important for the model
  df = df.drop(["Ticket", "Cabin", "Name"], axis=1)
  # Fill in the gaps in age with the average value
  df['Age'] = df['Age'].fillna(df['Age'].mean())
  return df

if __name__ == "__main__":
  url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
  data = clean_titanic_data(url)
  # Save the clean file to the data folder
  data.to_csv('data/cleaned_titanic.csv', index = False)
  print("Done! Clean data saved to the data/cleaned_titanic.csv folder.")
```
